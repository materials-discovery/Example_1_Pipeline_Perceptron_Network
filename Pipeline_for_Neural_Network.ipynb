```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Creating initial Data:

```python
T = np.array([-0.960, -0.577, -0.073, 0.377, 0.641, 0.660, 0.461, 0.134, -0.201, -0.434, -0.500, -0.393, -0.165, 0.099, 0.307, 0.396, 0.345, 0.182, -0.031, -0.219, -0.320])
X = np.linspace(-1,1,21)
X_bias = np.ones(21)
#   Creating a relational data set

X = np.concatenate([X_bias, X])
X = X.reshape(2,21).T
dataset = pd.DataFrame(X , columns={"Bias","X"})
dataset["T"] = T
```

# Print Data set

```python
dataset
```

```python
plt.plot(dataset, label='Inline Label')
plt.legend(["Bias","X:input","Target"])
```

Creating the parameters for the neural network (NOTE: For this example we use only the optimal values, usually findting such values require a training process)

```python
no_inputs = 1
no_hidden_neurons = 5
no_outputs = 1
# Weights for layer 1
V1 = np.array([[-2.7110, 3.6204],[1.2214, 3.8180],[-0.7778, 3.5548],[2.1751, 3.0169],[2.9979, 3.6398]])
print(V1.shape)
print(V1.shape)
# Weights for layer 2
W2 = np.array([-1.0295, -0.6334, -1.2985, 0.8719, 0.5937, 0.9906 ])
print(W2.shape)
```

```python
# Model of the Neural Network (to be discussed in the tutorial)
def forward_prop(X, V1, W2):
    # FIRST LAYER 
    S1 = X.dot(V1.T)
    M = (np.exp(S1) - np.exp(-S1))/(np.exp(S1) + np.exp(-S1))
    # SECOND LAYER
    bias_2 = np.ones(M.shape[0]) # getting number of columns
    M1 = np.column_stack( (bias_2, M) )
    Y = M1.dot(W2) 
    return Y, M
```

```python
Y, M = forward_prop(X, V1, W2)
print(Y, Y.shape)
plt.plot(Y, 'red', marker="o", label='Inline Label')  
plt.plot(T, 'b', marker="x", label='Inline Label') 
plt.legend(["Network Output","Desired target"])
plt.show()
```
